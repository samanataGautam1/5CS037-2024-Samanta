```python
import numpy as np

x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

add_xy = x + y
print("Task 1: Add the two arrays\n", add_xy)


subtract_xy = x - y
print("\nTask 2: Subtract the two arrays\n", subtract_xy)


multiply_x = x * 3
print("\nTask 3: Multiply the array with an integer\n", multiply_x)


square_x = np.square(x)
print("\nTask 4: Square of each element of the array\n", square_x)

dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)
print("\nTask 5: Dot products")
print("Dot product of v and w:", dot_vw)
print("Dot product of x and v:\n", dot_xv)
print("Dot product of x and y:\n", dot_xy)


concat_xy_row = np.concatenate((x, y), axis=0)
concat_vw_col = np.vstack((v, w))
print("\nTask 6: Concatenation")
print("Concatenate x and y along rows:\n", concat_xy_row)
print("Concatenate v and w along columns:\n", concat_vw_col)


print("\nTask 7: Attempt to concatenate x and v")
try:
    concat_xv = np.concatenate((x, v), axis=0)
    print(concat_xv)
except ValueError as e:
    print("Error:", e)

    v_reshaped = v.reshape(1, -1)
    concat_xv_fixed = np.concatenate((x, v_reshaped), axis=0)
    print("After fixing dimensions:\n", concat_xv_fixed)
```

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])


A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("Task 1: A * A^(-1) = I (Identity matrix)\n", I)


AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nTask 2: Prove AB != BA")
print("AB = \n", AB)
print("BA = \n", BA)


AB_T = np.transpose(AB)
B_T_A_T = np.dot(np.transpose(B), np.transpose(A))
print("\nTask 3: Prove (AB)^T = B^T * A^T")
print("Transpose of AB = \n", AB_T)
print("B^T * A^T = \n", B_T_A_T)


A_eq = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

A_inv_eq = np.linalg.inv(A_eq)
X = np.dot(A_inv_eq, B_eq)
print("\nTask 4: Solve the system of equations using inverse method")
print("Solution using inverse method: X = ", X)

X_solve = np.linalg.solve(A_eq, B_eq)
print("\nSolution using np.linalg.solve(): X = ", X_solve)
```
